```r
library("tidyverse")
```

```r
res <- read_csv("Tasklab2.csv")
```

```r
res
```

```r
avg_dplyr <- res %>% 
    group_by(dimension, operation) %>%
    summarise(
        time_sd = sd(times),
        times = mean(times)
    )  %>% 
    ungroup() %>%
    arrange(operation, dimension)
```

```r
avg_dplyr
```

res %>% 
    group_by(dimension, operation) %>%
    summarise(
        times = mean(times),
        times_sd = sd(times) # sd jest liczone dla jednego wiersza po agregracji, stad NA w wynikach
    )  %>% 
    ungroup() %>%
    arrange(operation, dimension)

```r
ggplot(avg_dplyr , aes(dimension,times,color=operation)) + geom_point() + xlab("Wymiary kolejnych macierzy/wektorow") +
    ylab("Osiagniete czasy mnozenia wektorow/macierzy w s") + geom_errorbar(aes(ymin = times - time_sd, ymax = times + time_sd, width=.25)) +
    ggtitle("Osiagniete czasy mnozenia macierzy/wektorow")
```

```r
avg_dplyr
```

```r
newdata <- avg_dplyr[c(1:500),c(1,4)]
```

```r
newdata
```

```r
x <- newdata[[1]]
y <- newdata[[2]]
mydata <- data.frame(x,y)
mydata
```

```r
fit <- lm(y ~ poly(x, 3, raw=TRUE), data=mydata)
```

```r
ggplot(newdata, aes(dimension,times)) + geom_point() + xlab("Wymiary kolejnych macierzy") + 
    ylab("Osiagniete czasy mnozenia macierzy w s") + ggtitle("Osiagniete czasy mnozenia macierzy")
```

```r
newer = data.frame(x = seq(1, 500, length.out=250))
```

```r
newer$y = predict(fit, newer)
```

```r
last_plot() + geom_line(data=newer, aes(x,y)) 
```
